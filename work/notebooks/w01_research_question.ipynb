## 2. The question: decision, action, cost of a wrong call
**Question:** Which pages should a content reviewer look at first for a refresh?

**Unit:** one page (content_id).
Decision: which pages get limited review time first.

**Action:** reviewer opens top-ranked pages, decides refresh/expand/prune/monitor.

**Cost of wrong call:** false positive = wasted review time. False negative = a real declining page keeps losing traffic unnoticed (worse).

**Frame:** For a content team deciding which pages to review first, we'll build a ranked priority queue using observed signals (staleness, CTR, position, engagement), scored by precision@K. A plain rule isn't enough. Week 2 showed a hand rule and a simple tree disagree depending on the metric. Claims will be observed/directional/decision-support only.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.

stale_visible = ((df["days_since_last_update"] >= 180) & (df["impressions_90d"] >= 500)).sum()
print(f"Stale + Visible Pages: {stale_visible} of {len(df)}")
```

## 3. Quick look at the data (2-3 real numbers)

Real numbers supporting this lane:

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.

declining_rate = (df["trend_direction"] == "down").mean()
stale_rate = (df["days_since_last_update"] >= 180).mean()
visible_declining = df[(df["trend_direction"] == "down") & (df["impressions_90d"] >= 100)].shape[0]

print(f"1. Declining pages: {declining_rate:.1%}")
print(f"2. Stale pages: {stale_rate:.1%}")
print(f"3. Declining + real demand (impressions>=100): {visible_declining}")
```

## 4. Careful words: what I can and can't claim

Can say: observed decline/CTR/staleness signals, ranked by strength — decision-support, not proof.
Can't say: that a refresh causes recovery, that I found a Google ranking factor, or that decline is certain vs. just "signals consistent with decline."

Note: is_declining_label comes from trend_direction (a rule, not a future outcome).
